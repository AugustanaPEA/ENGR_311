## Stress Transformation

As mentioned in the lecture, stress transformation is not something that is generally done by hand.  So this assigment asks you to build a little code that takes an initial expression for a state of stress, and produces a transformed expression of that same state. <br><br>

Your inputs should be the initial values for the normal stresses in the x and y direction, the shear stress, and the rotation.  Be sure to follow sign convention: <br><br>

* Compressive stresses are negative, while tensile stresses are positive
* Shear is positive if it is CCW on x-side of the element
* Rotation is positive if it is CCW

Here's a little help getting started.  The first cell gives you a structure for the function you'll create.   The second runs the function and prints the output.

```python
import numpy as np

def transform(sx,sy,tau,rot):
    rot_r = (rot/180.0)*np.pi
    sxp = (sx + sy)/2.0 + ((sx-sy)/2.0)*np.cos(2.0*rot_r) + tau*np.sin(2.0*rot_r)
    syp = (sx + sy)/2.0 - ((sx-sy)/2.0)*np.cos(2.0*rot_r) - tau*np.sin(2.0*rot_r)
    taup = -((sx-sy)/2.0)*np.sin(2.0*rot_r) + tau*np.cos(2.0*rot_r)
    return sxp, syp, taup
```

```python
# Units must all be the same (MPa, GPa, etc...)
# Rotation is in degrees
sx = 45
sy = -60
tau = 30
rot = -10

sxp, syp, taup =transform(sx,sy,tau,rot)

print("sigma_x = ", round(sxp,2))
print("sigma_y = ", round(syp,2))
print(" shear = ", round(taup,2))
```
